```python
import geopandas as gpd
import pandas as pd
import glob
```

```python
arquivos = glob.glob('resultados/parciais/*.gpkg')
```

```python
count = 0
gdfs = [] 

for arquivo in arquivos:
    
    print(arquivo, end=' ')

    gdf = gpd.read_file(arquivo).loc[:, ['sqlc', 'BHM_min', 'BHM_max', 'BHM_count', 'BHM_sum']]
    gdfs.append(gdf)

    count += 1
    print(count)
    # if count > 20:
    #     break
```

```python
gdf_lotes_lidar = pd.concat(gdfs, ignore_index=True).fillna(0.0)
```

```python
gdf_lotes_lidar
```

```python
gdf_grouped = gdf_lotes_lidar.groupby('sqlc').agg(
    {
        'BHM_min':'min',
        'BHM_max':'max',
        'BHM_count':'sum',
        'BHM_sum':'sum'
    }
).reset_index()
```
