```python
import geopandas as gpd
import pandas as pd
import numpy as np
from rasterstats import zonal_stats
```

```python
gdf_lotes = gpd.read_file('resultados/lotes.gpkg')
```

```python
gdf_quadriculas = gpd.read_file('zip://downloads/SIRGAS_SHP_quadriculamdt.zip!/SIRGAS_SHP_quadriculamdt/')
_ = gdf_quadriculas.set_crs(epsg=31983, inplace=True)
```

```python
count = 1

for quadricula in gdf_quadriculas.itertuples():
    print(quadricula.qmdt_cod)
    print(f'{count}/{gdf_quadriculas.shape[0]}')
    count += 1
    clipe = gpd.clip(gdf_lotes, gdf_quadriculas[gdf_quadriculas.loc[:, 'qmdt_cod'] == quadricula.qmdt_cod])
    if clipe.shape[0] == 0:
        continue

    zonal = zonal_stats(clipe, 
                        f'../sampa-lidar/results/2020/BHM/BHM-{quadricula.qmdt_cod}-2020-1m.tiff',
                        stats=['max', 'min', 'count', 'sum', 'mean'])
    df_lote_bhm = pd.concat([clipe.reset_index(), pd.DataFrame(zonal).reset_index().add_prefix('BHM_')], axis=1)
    df_lote_bhm.drop(columns=['index', 'index']).to_file(f'resultados/parciais/lotes-{quadricula.qmdt_cod}-2020-1m.gpkg', driver='GPKG')
    # break
```
